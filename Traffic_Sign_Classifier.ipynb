### Model Architecture

- Setup TenserFlow

````python
### Define your architecture here.
### Feel free to use as many code cells as needed.
import tensorflow as tf

# Arguments used for tf.truncated_normal, randomly defines variables for the weights and biases for each layer
mu = 0
sigma = 0.1
````

- LeNet fucntion

````python
def conv2d(input, weight_shape, strides = [1,1,1,1], padding = 'VALID'):
    # Filter (weights and bias)
    F_W = tf.Variable(tf.truncated_normal(shape = weight_shape, mean = mu, stddev = sigma))
    F_b = tf.Variable(tf.zeros(weight_shape[-1]))
    return tf.nn.conv2d(input, F_W, strides, padding) + F_b
````

````python
def max_pool(input):
    return tf.nn.max_pool(input, ksize = [1, 2, 2, 1], strides = [1, 2, 2, 1], padding = 'VALID')
````

````python
def fully_connected(input, weight_shape):
    F_W = tf.Variable(tf.truncated_normal(shape = weight_shape, mean = mu, stddev = sigma))
    F_b = tf.Variable(tf.zeros(weight_shape[-1]))
    return tf.matmul(input, F_W) + F_b
````

````python
from tensorflow.contrib.layers import flatten

def LeNet(x):    

    # Layer 1: Convolutional. Input = 32x32x1. Output = 28x28x6.
    conv1 = conv2d(input = x, weight_shape = (5, 5, 3, 6))
    
    # Activation.
    conv1 = tf.nn.relu(conv1)
    
    # Pooling. Input = 28x28x6. Output = 14x14x6.
    conv1 = max_pool(conv1)
    
    # Layer 2: Convolutional. Output = 10x10x16.
    conv2 = conv2d(input = conv1, weight_shape = (5, 5, 6, 16))
    
    # Activation.
    conv2 = tf.nn.relu(conv2)
    
    #  Pooling. Input = 10x10x16. Output = 5x5x16.
    conv2 = max_pool(conv2)
    
    # Flatten. Input = 5x5x16. Output = 400.
    fc0 = flatten(conv2)
    
    # Layer 3: Fully Connected. Input = 400. Output = 120.
    fc1 = fully_connected(input = fc0, weight_shape = (400, 120))
    
    #  Activation.
    fc1 = tf.nn.relu(fc1)
    
    # Layer 4: Fully Connected. Input = 120. Output = 84.
    fc2 = fully_connected(input = fc1, weight_shape = (120, 84))
    
    #  Activation.
    fc2 = tf.nn.relu(fc2)
    
    # Layer 5: Fully Connected. Input = 84. Output = 43.
    logits = fully_connected(input = fc2, weight_shape = (84, 43))
   
    return logits
````

### Train, Validate and Test the Model

A validation set can be used to assess how well the model is performing. A low accuracy on the training and validation
sets imply underfitting. A high accuracy on the training set but low accuracy on the validation set implies overfitting.

- Training Pipeline

````python
# set learning rate later
rate = tf.placeholder(tf.float32)

logits = LeNet(x)
cross_entropy = tf.nn.softmax_cross_entropy_with_logits(labels=one_hot_y, logits=logits)
loss_operation = tf.reduce_mean(cross_entropy)
optimizer = tf.train.AdamOptimizer(learning_rate = rate)
training_operation = optimizer.minimize(loss_operation)
````

- Model Evaluation

````python
correct_prediction = tf.equal(tf.argmax(logits, 1), tf.argmax(one_hot_y, 1))
accuracy_operation = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
saver = tf.train.Saver()

def evaluate(X_data, y_data):
    num_examples = len(X_data)
    total_accuracy = 0
#     sess = tf.get_default_session()
    for offset in range(0, num_examples, BATCH_SIZE):
        batch_x, batch_y = X_data[offset:offset+BATCH_SIZE], y_data[offset:offset+BATCH_SIZE]
        accuracy = sess.run(accuracy_operation, feed_dict={x: batch_x, y: batch_y})
        total_accuracy += (accuracy * len(batch_x))
    return total_accuracy / num_examples
````

- Train The Model

````python
EPOCHS = 20
BATCH_SIZE = 128

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    num_examples = len(X_train)
    
    dynamic_learning_rate = 0.001 # start learning rate with 0.001
    validation_accuracies = []
    print("Training...")
    print()
    for i in range(EPOCHS):
        X_train, y_train = shuffle(X_train, y_train)
        for offset in range(0, num_examples, BATCH_SIZE):
            end = offset + BATCH_SIZE
            batch_x, batch_y = X_train[offset:end], y_train[offset:end]
            sess.run(training_operation, feed_dict={x: batch_x, y: batch_y, rate: dynamic_learning_rate})
            
        validation_accuracy = evaluate(X_valid, y_valid)
        validation_accuracies.append(validation_accuracy)    
        print("EPOCH {}, rate = {:.4f}, Validation Accuracy = {:.3f}".format(i+1, dynamic_learning_rate, validation_accuracy))
        print()
        # dynamicly adjust learning rate based on the current accuracy result
        if validation_accuracy > 0.975:
            break
        elif validation_accuracy > 0.97:
            dynamic_learning_rate = 0.0002
        elif validation_accuracy > 0.95: 
            dynamic_learning_rate = 0.0004
        elif validation_accuracy > 0.92: 
            dynamic_learning_rate = 0.0006
        elif validation_accuracy > 0.9: 
            dynamic_learning_rate = 0.0008
        elif validation_accuracy > 0.88: 
            dynamic_learning_rate = 0.0009
    
    saver.save(sess, './lenet')
    print("Model saved")
````
